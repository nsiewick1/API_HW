Example Output:

```python
# Importing libraries
import json
import pandas as pd
import json
import requests
```

```python
# Prompting the user to input two or more stock symbols
stock=input()
```

```python
# Using the API to fetch data for the given symbols
# apikey = vM8wGFsXlW9QxYaeMJQxI2W8bUMXO9Wja1ZY5GTw
apikey=""

url = "https://yfapi.net/v6/finance/quote"

querystring = {"symbols": stock}

headers = {
    'x-api-key': apikey
    }

response = requests.get(url, headers=headers, params=querystring)

print(response.text)
```

```python
# Clean up output format
print(json.dumps(response.json(), indent=4))
```

```python
# Put in dataframe
df = pd.DataFrame(response.json()['quoteResponse']['result'])
print(df[["symbol","longName","regularMarketPrice"]])
```

```python
# Example Output Format
df.rename(columns={"symbol": "Stock Ticker", "longName": "Company", "regularMarketPrice": "Current Market Price"}, inplace=True)

example_output_format = "\n\n"

for index, row in df.iterrows():
    example_output_format += f"Stock Ticker: {row['Stock Ticker']}, Company: {row['Company']}, Current Market Price: ${row['Current Market Price']:.2f}\n"

print(example_output_format)
```

Task 2.1: Fetch Additional Data Using Modules 10pts
- Allow the user to choose a module from the Quote Summary Endpoint.

Find:
52 Week High
52 Week Low
Return on Assets (ROA)
- Make an API request using the chosen module.
- Convert the response into a Pandas DataFrame.
- Display the DataFrame.

```python
# Prompting the user to input a module from the Quote Summary Endpoint. Find: 52 Week High 52 Week Low Return on Assests (ROA)
module = input()
```

```python
# Make an API request using the chosen module.
apikey= ""

url = "https://yfapi.net/v11/finance/quoteSummary/AAPL"

querystring = {
    "modules": module
}

headers = {
    'x-api-key': apikey
}

# headers = {'Authorization': f'Bearer {apikey}'}

response = requests.get(url, headers=headers, params=querystring)

# Pretty print the JSON response
print(json.dumps(response.json(), indent=4))
```

```python
# Put in dataframe
# Used ChatGPT to help me fix syntax, merging dataframes, and subsetting certain information from merged dataframe error
df1 = pd.DataFrame(response.json()['quoteSummary']['result'])
df_summary = pd.json_normalize(df1['summaryDetail'])
df_financial = pd.json_normalize(df1['financialData'])
df_merged = pd.concat([df_summary, df_financial], axis=1)
df_merged.rename(columns={"fiftyTwoWeekHigh.raw": "52 Week High", "fiftyTwoWeekLow.raw": "52 Week Low", "returnOnAssets.raw": "Return On Assets (ROA)"}, inplace=True)
print(df_merged[["52 Week High","52 Week Low", "Return On Assets (ROA)"]])
```

Task 2.2: Fetch Additional Data Using Modules 10pts

- Find Current Trending Stocks , Their Actual Name and Ticker and Current Price as well as
their 52 high and low.

```python
# Find Current Trending Stocks
apikey=""

url = "https://yfapi.net/v1/finance/trending/US"

headers = {
    'x-api-key': apikey
}

# headers = {'Authorization': f'Bearer {apikey}'}

response = requests.get(url, headers=headers, params=querystring)

# Pretty print the JSON response
print(json.dumps(response.json(), indent=4))
```

```python
# Get trending stocks and extract tickers directly
# Used ChatGPT to help me understand how to formulate syntax to get trending stocks and extract tickers directly
trending_stocks = [stock['symbol'] for stock in response.json()['finance']['result'][0]['quotes']]

# List to store stock DataFrames
stock_dataframes = []

# For-loop to get specified information in instructions for each trending stock symbol
for symbol in trending_stocks:
    stock_url = f"https://yfapi.net/v11/finance/quoteSummary/{symbol}"
    querystring = {"modules": "summaryDetail,price"}

    stock_data = requests.get(stock_url, headers=headers, params=querystring).json()
    df_1 = pd.DataFrame(stock_data['quoteSummary']['result'])

    # Extract and merge price & summaryDetail data
    df_sum = pd.json_normalize(df_1['summaryDetail'])
    df_price = pd.json_normalize(df_1['price'])
    df_merge = pd.concat([df_sum, df_price], axis=1)

    # Rename columns
    df_merge.rename(columns={
            "fiftyTwoWeekHigh.raw": "52 Week High",
            "fiftyTwoWeekLow.raw": "52 Week Low",
            "regularMarketPrice.raw": "Current Price",
            "symbol": "Stock Ticker",
            "longName": "Company Name"}, inplace=True)

    # Put relevant columns to list of dataframes
    stock_dataframes.append(df_merge[["Stock Ticker", "Company Name", "Current Price", "52 Week High", "52 Week Low"]])

# Combine all stock data
df_final = pd.concat(stock_dataframes, ignore_index=True)
print(df_final)
```
